```python
#LEVEL 2
#Replace the missing values with the average of that column (Do this for the column of evaporation only)
```

```python
Average_of_evaporation=weather['Evaporation'].mean(skipna=True)
print(Average_of_evaporation)

```

```python
weather['Evaporation'].fillna(Average_of_evaporation, inplace = True)
  
```
